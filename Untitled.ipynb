```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
cols = ['Health Level', 'Hearts Available', 'Weeks on waitlist', 'Action']
```

```python
data = pd.read_csv('./state_opt_action.csv', names=cols)
```

```python
data['Action'] = data['Action'].apply(lambda value: 'Transplant' if value == 1 else 'Wait')
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

xs = data['Health Level']
ys = data['Hearts Available']
zs = data['Weeks on waitlist']
ax.scatter(xs, ys, zs, s=50, alpha=0.6, edgecolors='w')

ax.set_xlabel('Health Level')
ax.set_ylabel('Hearts Available')
ax.set_zlabel('Weeks on Waitlist')
```

```python
a = sns.countplot(data=data, x='Hearts Available',hue="Action")
```

```python
b = sns.countplot(data=data, x='Health Level',hue="Action")
```

```python
df = data.copy()
df['Weeks on waitlist'] = df['Weeks on waitlist'].apply(lambda value: '0-10'
                                                        if value <= 10 else '11-20' 
                                                        if value <= 20 else '21-30'
                                                        if value <= 30 else '31-40'
                                                        if value <= 40 else '41-50'
                                                        if value <= 50 else '51-60'
                                                        if value <= 60 else '61-70'
                                                        if value <= 70 else '71-80'
                                                        if value <= 80 else '81-90')


c = sns.countplot(data=df, x= 'Weeks on waitlist',hue="Action")
```

```python
pp = sns.pairplot(df, hue='Action', size=1.8, aspect=1.8, 
                  palette={"Transplant": "#FF9999", "Wait": "#FFE888"},
                  plot_kws=dict(edgecolor="black", linewidth=0.5))
fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.3)
t = fig.suptitle('Pairwise Plots', fontsize=14)
```
